```python
# Connection and data frame creation
import mysql.connector
import pandas as pd

connection = mysql.connector.connect(
user='root', 
password='12345@As', 
host='localhost', 
database='ecommerce'
)
# Create a cursor object to execute SQL queries 
cursor = connection.cursor()
# Query data from the 'customer' table
cursor.execute('SELECT * FROM customer')
#After fetching data from the database we are storing it into Pandas DataFrame 
customer_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
# Query data from the 'product' table
cursor.execute('SELECT * FROM product')
product_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
# Query data from the 'order_details' table
cursor.execute('SELECT * FROM order_details')
order_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
connection.close()
```

```python
# 1. Aggregate the total revenue for each product
product_revenue = order_data.groupby('product_id')['total_price'].sum().reset_index()

# 2. Merge with product_data to get product names
product_revenue = pd.merge(product_revenue, product_data[['product_id', 'product_name']], on='product_id')

# 3. Plotting the total revenue for each product
plt.figure(figsize=(12, 6))
plt.bar(product_revenue['product_name'], product_revenue['total_price'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Generated from Orders by Product')
plt.xticks(rotation=45, ha='right')  # Rotate product names for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
# 1. Aggregate the total revenue for each product category
category_revenue = order_data.merge(product_data[['product_id', 'category']], on='product_id')
category_revenue = category_revenue.groupby('category')['total_price'].sum().reset_index()

# 2. Calculate percentage of total revenue for each category
total_revenue = category_revenue['total_price'].sum()
category_revenue['percentage'] = (category_revenue['total_price'] / total_revenue) * 100

# 3. Plotting the revenue percentage for each category
plt.figure(figsize=(8, 8))
plt.pie(category_revenue['percentage'], labels=category_revenue['category'], autopct='%1.1f%%', colors=plt.cm.Paired(range(len(category_revenue))))
plt.title('Revenue Percentage by Product Category')
plt.show()
```

```python
# 1. Merge order_data with product_data to get original and selling prices
merged_data = order_data.merge(product_data[['product_id', 'original_price', 'selling_price']], on='product_id')

# 2. Calculate profit per unit and total profit for each product
merged_data['profit_per_unit'] = merged_data['selling_price'] - merged_data['original_price']
merged_data['total_profit'] = merged_data['profit_per_unit'] * merged_data['quantity']

# 3. Aggregate total profit for each product
product_profit = merged_data.groupby('product_id').agg({'total_profit': 'sum'}).reset_index()

# 4. Merge with product_data to get product names
product_profit = pd.merge(product_profit, product_data[['product_id', 'product_name']], on='product_id')

# 5. Sort by total profit and get top products (e.g., top 10)
top_products_by_profit = product_profit.sort_values(by='total_profit', ascending=False).head(10)

# 6. Plotting the top products by total profit
plt.figure(figsize=(12, 6))
plt.bar(top_products_by_profit['product_name'], top_products_by_profit['total_profit'], color='green')
plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.title('Top Products by Total Profit')
plt.xticks(rotation=45, ha='right')  # Rotate product names for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
# 1. Aggregate the total order quantity for each product
product_quantities = order_data.groupby('product_id')['quantity'].sum().reset_index()

# 2. Merge with product_data to get product names
product_quantities = pd.merge(product_quantities, product_data[['product_id', 'product_name']], on='product_id')

# 3. Find the product with the highest and lowest order quantities
highest_quantity_product = product_quantities.loc[product_quantities['quantity'].idxmax()]
lowest_quantity_product = product_quantities.loc[product_quantities['quantity'].idxmin()]

# 4. Plotting the products with highest and lowest order quantities
plt.figure(figsize=(10, 5))
plt.bar(['Highest Quantity', 'Lowest Quantity'], [highest_quantity_product['quantity'], lowest_quantity_product['quantity']], color=['blue', 'red'])
plt.xlabel('Product')
plt.ylabel('Order Quantity')
plt.title('Product with Highest and Lowest Order Quantities')
plt.xticks(['Highest Quantity', 'Lowest Quantity'], [highest_quantity_product['product_name'], lowest_quantity_product['product_name']])
plt.tight_layout()
plt.show()
```

```python
# 1. Aggregate the total order quantity for each customer
customer_quantities = order_data.groupby('customer_id')['quantity'].sum().reset_index()

# 2. Merge with customer_data to get customer names
customer_quantities = pd.merge(customer_quantities, customer_data[['customer_id', 'name']], on='customer_id')

# 3. Find the customer with the highest and lowest order quantities
highest_quantity_customer = customer_quantities.loc[customer_quantities['quantity'].idxmax()]
lowest_quantity_customer = customer_quantities.loc[customer_quantities['quantity'].idxmin()]

# 4. Plotting the customers with highest and lowest order quantities
plt.figure(figsize=(10, 5))
plt.bar(['Highest Quantity', 'Lowest Quantity'], [highest_quantity_customer['quantity'], lowest_quantity_customer['quantity']], color=['blue', 'red'])
plt.xlabel('Customer')
plt.ylabel('Order Quantity')
plt.title('Customers with Highest and Lowest Order Quantities')
plt.xticks(['Highest Quantity', 'Lowest Quantity'], [highest_quantity_customer['name'], lowest_quantity_customer['name']])
plt.tight_layout()
plt.show()
```

```python
# 1. Convert 'order_date' to datetime format
order_data['order_date'] = pd.to_datetime(order_data['order_date'])

# 2. Extract month and year from 'order_date'
order_data['month_year'] = order_data['order_date'].dt.to_period('M')

# 3. Aggregate total sales by month and year
monthly_sales = order_data.groupby('month_year')['total_price'].sum().reset_index()

# 4. Plotting the month-wise total sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['month_year'].astype(str), monthly_sales['total_price'], marker='o', color='b')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.title('Month-wise Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. Convert 'order_date' to datetime format
order_data['order_date'] = pd.to_datetime(order_data['order_date'])

# 2. Aggregate order counts by date
daily_order_counts = order_data['order_date'].value_counts().sort_index()

# 3. Find the peak order date
peak_order_date = daily_order_counts.idxmax()
peak_order_count = daily_order_counts.max()

# 4. Plotting the order counts with a marker for the peak date
plt.figure(figsize=(12, 6))
plt.plot(daily_order_counts.index, daily_order_counts.values, marker='o', color='b')
plt.axvline(x=peak_order_date, color='r', linestyle='--', label=f'Peak Order Date: {peak_order_date.date()}')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.title('Daily Order Counts with Peak Order Date')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
