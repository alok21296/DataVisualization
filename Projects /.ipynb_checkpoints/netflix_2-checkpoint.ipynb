```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("netflixdata.csv")

# Display the last 8,000 rows
df.tail(8000)
```

```python
# Display basic information about the DataFrame
print("Basic Information:")
df.info()
```

```python
# Display the first few rows of the DataFrame
print("\nFirst Few Rows:")
df.head()
```

```python
# see if we have null values in dataset
df.isna().sum() #on rating we have 4 rows with null values, and on duration 3
```

```python
df['director'] = df['director'].fillna('Unspecified')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['duration']= df['duration'].fillna(df['duration'].mode()[0])
#let's replace that "s" in the show_id
df['show_id'] = df['show_id'].str.replace('s','')
```

```python
df.isna().sum()
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Content Type Analysis
content_type_counts = df['type'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(content_type_counts.index, content_type_counts.values, color=['blue', 'orange'])
plt.title('Distribution of Content Types')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
# Genre Analysis
df['genres'] = df['listed_in'].str.split(', ')
all_genres = df.explode('genres')
genre_counts = all_genres['genres'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.barh(genre_counts.index, genre_counts.values, color='green')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
# Content Release Trends
release_trends = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(release_trends.index, release_trends.values, marker='o', color='teal')
plt.title('Content Release Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python
# Country Analysis
df['countries'] = df['country'].str.split(', ')
all_countries = df.explode('countries')
country_counts = all_countries['countries'].value_counts().head(10)
plt.figure(figsize=(12, 6))
plt.barh(country_counts.index, country_counts.values, color='purple')
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Rating Analysis
rating_counts = df['rating'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(rating_counts.index, rating_counts.values, color='red')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#Top Directors analysis 
titles_per_director = df.groupby('director')['show_id'].count().astype(int).reset_index()
titles_per_director.columns = ['Director','Titles No']

director_titles = titles_per_director.sort_values(['Titles No','Director'], ascending=False).iloc[1:6]
director_titles = director_titles.sort_values('Titles No', ascending = True)
plt.barh(director_titles['Director'], director_titles['Titles No'], color = '#D00202')
plt.title('Top 5 directors by number of Titles')
plt.ylabel('Director')
plt.xlabel('Number of Titles')


for index,value in enumerate(director_titles['Titles No']):
    plt.text(value, index, str(value), ha='left', va='center', fontsize= 9, fontweight='semibold')

plt.show()
```

```python
 # Insights and Conclusion
print("\nInsights and Conclusions:")
print("- Netflix has more movies than TV shows.")
print("- The top genres include Drama, Comedy, and Documentaries.")
print("- There has been a steady increase in content releases over the years.")
print("- The United States is the largest producer of content for Netflix.")
print("- The most common ratings are TV-MA and TV-14.")
```
