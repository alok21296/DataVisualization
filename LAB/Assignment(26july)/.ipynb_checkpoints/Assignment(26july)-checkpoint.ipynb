```python
#LAB 2
#Ques-1-Convert the given numpy array then display the array
#Input: my_list=[1,2,3,4,5]
#Output:NumPy array:=
import numpy as np
my_list=[1,2,3,4,5]
print("Input:",my_list)
a=np.array(my_list)
print("Output",a)
```

```python
#LAB 3
#1-Write a numpy program to create an array of 10 zeroes,10 ones and 10 fives.
zeroes=np.zeros(10,dtype=int)
ones=np.ones(10,dtype=int)
fives=np.full(10,5,dtype=int)
print(zeroes)
print(ones)
print(fives)
```
