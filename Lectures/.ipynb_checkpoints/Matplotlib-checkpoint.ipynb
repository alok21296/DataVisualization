```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80,90]
y=[1,2,3,4,5,6,7,8,9]
plt.subplot(2,2,1)
plt.plot(x,y,color="g")

plt.subplot(2,2,2)
plt.pie([1],colors="y")

plt.subplot(2,2,3)
a=[20,40,60]
plt.pie(a)

plt.subplot(2,2,4)
a1=["r","a","h","u"]
a2=[20,40,60,80]
plt.bar(a1,a2)

plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 5, 11]
# Creating plot
plt.plot(x, y)
# Adding title and axis Labels
plt. title( 'Plot with General Text') 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Adding general text
plt. text(2, 8, 'General Text', fontsize=10, color='green')
plt. annotate(r'Java', xy=(4, 5), xytext=(3, 10), arrowprops=dict (facecolor='red' , shrink=0.50), fontsize=20, color='red')
plt. legend(["Java"],loc=5,facecolor="yellow", edgecolor="red" , framealpha=0.5, shadow=True)
# Showing plot
plt.show( )
```
